```python
import numpy as np
import pandas as pd
inputs = open("input.txt")
n = int(inputs.readline())
result = open("output.txt", "w")

def getListof():
    arr1 = list(inputs.readline().replace("[", "").replace("]", "").split(","))
    arr1 = [int(a) for a in arr1]
    return arr1

for i in range(0, n, 1):
    takeOutOrders = getListof()
    dineInOrders = getListof()
    servedOrders = getListof()
    temp1 = 0
    temp2 = 0
    validate = 1
    for j in range(0, len(servedOrders), 1):
        if temp1 < len(takeOutOrders) and takeOutOrders[temp1]==servedOrders[j]:
            temp1 += 1
        elif temp2 < len(dineInOrders) and dineInOrders[temp2]==servedOrders[j]:
            temp2 += 1
        else:
            validate = 0
            break
    if validate == 1:
        result.write("Valid\n")
    else:
        result.write("Invalid\n")
```
